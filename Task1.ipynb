2.	Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.


```python
data = np.random.exponential(scale=1.0, size=(100000, 1))

plt.hist(data, bins=100, density=True, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), bins=100, density=True, histtype="step", color="red", label="Uniform")
plt.hist(np.random.randn(100000), bins=100, density=True, histtype="step", color="green", label="Normal")

plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3.	Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D plot of Z = X^2 + Y^2')
plt.show()
```
